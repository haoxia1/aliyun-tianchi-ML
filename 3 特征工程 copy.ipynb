# 特征工程

1. 异常值分析

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats # 统计
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data_file = './zhengqi_train.txt' # ./代表当前目录
test_data_file = './zhengqi_test.txt'
train_data = pd.read_csv(train_data_file, sep='\t', encoding='utf-8') # 分隔符为'\t'
test_data = pd.read_csv(test_data_file, sep='\t', encoding='utf-8')
```

```python
# 如V9是异常值，分别将训练集和测试集中的异常值删除
train_data = train_data[train_data['V9'] > -7.5] # 2888行变成2886行 删除了俩异常值
test_data = test_data[test_data['V9'] > -7.5] # 1925行不变
```

2. 最大值和最小值的归一化

```python
from sklearn import preprocessing 

feature_columns = train_data.columns.tolist()[:-1]

min_max_scaler = preprocessing.MinMaxScaler()  # 创建对象
train_data1 = min_max_scaler.fit_transform(train_data[feature_columns]) # 调用对象方法
test_data1 = min_max_scaler.fit_transform(test_data)
train_data1 = pd.DataFrame(train_data1) # ndarray -> df
train_data1.columns = feature_columns # add columns
train_data1['target'] = train_data['target'] # add 'target'

test_data1 = pd.DataFrame(test_data1)
test_data1.columns = feature_columns

train_data = train_data1
test_data = test_data1


train_data = train_data.fillna(0)
test_data = test_data.fillna(0)

display(train_data)
```

```python

from sklearn import preprocessing 

features_columns = [col for col in train_data.columns if col not in ['target']]

min_max_scaler = preprocessing.MinMaxScaler()

min_max_scaler = min_max_scaler.fit(train_data[features_columns])

train_data_scaler = min_max_scaler.transform(train_data[features_columns])
test_data_scaler = min_max_scaler.transform(test_data[features_columns])

train_data_scaler = pd.DataFrame(train_data_scaler)
train_data_scaler.columns = features_columns

test_data_scaler = pd.DataFrame(test_data_scaler)
test_data_scaler.columns = features_columns

train_data_scaler['target'] = train_data['target']
display(train_data_scaler.describe())
display(test_data_scaler.describe())
```

7. PCA处理

```python
# PCA可去除数据的多重共线性，并进行降维；

# PCA将38个特征，降维成了21个主成分
from sklearn.decomposition import PCA # 主成分分析法
```

```python
# PCA保留16个主成分(与上述方法等效吧)
pca = PCA(n_components=16)
new_train_pca_16 = pca.fit_transform(train_data_scaler.iloc[:,:-1])
new_test_pca_16 = pca.fit_transform(test_data_scaler)

new_train_pca_16 = pd.DataFrame(new_train_pca_16)
new_test_pca_16 = pd.DataFrame(new_test_pca_16)

new_train_pca_16['target'] = train_data_scaler['target']
train_data = new_train_pca_16
test_data = new_test_pca_16
```

```python
# 源数据 
from sklearn.linear_model import LinearRegression   # 线性回归
from sklearn.neighbors import KNeighborsRegressor   # KNN回归
from sklearn.tree import DecisionTreeRegressor      # 决策树回归
from sklearn.ensemble import RandomForestRegressor  # 随机森林回归
from sklearn.svm import SVR                         # 支持向量回归
import lightgbm as lgb                              # LightGBM模型


from sklearn.model_selection import train_test_split # 切分数据
from sklearn.metrics import mean_squared_error       # MSE


train = train_data[test_data.columns] # train 就是X，没有target (2888, 38)
target = train_data['target']  # (2888, )

# # 切分训练集 训练数据80%，验证数据20%； 测试集始终没动
train_data, test_data, train_target, test_target = train_test_split(
    train, target, test_size=0.2, random_state=0) # 那原来的train_data和test_data变量怎么办


# 1
clf = LinearRegression()
clf.fit(train_data, train_target)
test_pred = clf.predict(test_data)
score = mean_squared_error(test_target, clf.predict(test_data))  # MSE
print(f'LinearRegression_MSE:  {score:.3f}') 
print(f'LinearRegression_R2_score:  {clf.score(test_data, test_target):.3f}')

# 2
clf = KNeighborsRegressor(n_neighbors=3)  # 最近的3个
clf.fit(train_data, train_target)
test_pred = clf.predict(test_data)
score = mean_squared_error(test_target, clf.predict(test_data))
print(f'KNeighborsRegressor_MSE:  {score:.3f}') 
print(f'KNeighborsRegressor_R2_score:  {clf.score(test_data, test_target):.3f}')

# 3
clf = DecisionTreeRegressor()
clf.fit(train_data, train_target)
test_pred = clf.predict(test_data)
score = mean_squared_error(test_target, clf.predict(test_data))
print(f'DecisionTreeRegressor_MSE:  {score:.3f}')   # 这准确率不错啊
print(f'DecisionTreeRegressor_R2_score:  {clf.score(test_data, test_target):.3f}')

# 4
clf = RandomForestRegressor(n_estimators=200)  # 200棵树模型
clf.fit(train_data, train_target)
test_pred = clf.predict(test_data)
score = mean_squared_error(test_target, clf.predict(test_data))
print(f'RandomForestRegressor_MSE:  {score:.3f}')  # 准确率就这？？
print(f'RandomForestRegressor_R2_score:  {clf.score(test_data, test_target):.3f}')

# 5
clf = lgb.LGBMRegressor(
    learning_rate=0.01,
    max_depth=-1,
    n_estimators=5000,
    boosting_type='gbdt',
    random_state=2019,
    objective='regression'
)
clf.fit(X=train_data, y=train_target, eval_metric='MSE', verbose=50)
score = mean_squared_error(test_target, clf.predict(test_data))
print(f'lightGbm_MSE:  {score:.3f}')
print(f'lightGbm_R2_score:  {clf.score(test_data, test_target):.3f}')
```
